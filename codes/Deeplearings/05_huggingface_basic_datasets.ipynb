## 파일이나 pandas 를 Dataset 변환과정

```python
import pandas as pd

# 데이터 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 30, 18, 35, 28],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles'],
    'Score': [85, 90, 78, 88, 95]
}

df_data = pd.DataFrame(data)
df_data.head()
```
